# üî¨ Detecci√≥n de Im√°genes Generadas por IA
## Parte I: An√°lisis de Covarianza de Gradientes

Este notebook demuestra el m√©todo base de detecci√≥n utilizando las correlaciones entre gradientes de luminancia. El c√≥digo ha sido refactorizado al paquete `ai_detector`.

```python
%load_ext autoreload
%autoreload 2
import sys
import os
# A√±adir src al path para poder importar si no se ha instalado con pip -e .
sys.path.append(os.path.abspath('../src'))

import numpy as np
import matplotlib.pyplot as plt
from ai_detector.features.gradient import GradientCovarianceDetector
from ai_detector.data.generation import generate_natural_like_image, generate_synthetic_image
from ai_detector.visualization.plots import visualize_analysis

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['font.size'] = 11
```

## 1. Generaci√≥n de Datos de Prueba
Generamos una imagen "natural" (ruido 1/f) y una "sint√©tica" (con artefactos de grilla).

```python
# Generar im√°genes
img_natural = generate_natural_like_image(seed=42)
img_synthetic = generate_synthetic_image(seed=42)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(img_natural)
ax[0].set_title('Simulaci√≥n Natural (1/f)')
ax[0].axis('off')
ax[1].imshow(img_synthetic)
ax[1].set_title('Simulaci√≥n Sint√©tica (Artefactos)')
ax[1].axis('off');
```

## 2. An√°lisis con GradientCovarianceDetector
Utilizamos la clase refactorizada para analizar ambas im√°genes.

```python
detector = GradientCovarianceDetector()

print("Analizando imagen natural...")
res_natural = detector.analyze(img_natural)
visualize_analysis(img_natural, res_natural, "An√°lisis: Imagen Natural")

print("\nAnalizando imagen sint√©tica...")
res_synthetic = detector.analyze(img_synthetic)
visualize_analysis(img_synthetic, res_synthetic, "An√°lisis: Imagen Sint√©tica")
```

## 3. Comparaci√≥n de Features
Comparamos las m√©tricas extra√≠das.

```python
import pandas as pd

df = pd.DataFrame([
    res_natural['features'],
    res_synthetic['features']
], index=['Natural', 'Sint√©tica'])

display(df.T)
```
